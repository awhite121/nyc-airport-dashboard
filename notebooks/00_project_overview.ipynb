# NYC Taxi Trip Duration & Late-Risk Prediction

This notebook gives the clean project overview for the GitHub repo.

## What this project does

The goal is to estimate two things for Manhattan → JFK/LGA yellow taxi trips:

1. **Trip duration** — how long the ride will take.
2. **Late-trip risk** — whether the ride is unusually slow relative to similar trips.

The project combines regression, imbalanced classification, threshold tuning, and buffer-time guidance to turn historical taxi data into a practical airport departure advisor.


## Dataset Snapshot

| Item | Value |
|---|---:|
| Raw August 2025 taxi rows uploaded | 3,574,091 |
| Clean Manhattan → JFK/LGA trips | 43,079 |
| JFK trips | 18,986 |
| LGA trips | 24,093 |
| Late-trip rate | 18.1% |
| Median JFK duration | 54.2 min |
| Median LGA duration | 28.5 min |

The raw full taxi file is not included in the repo because it is too large for normal GitHub usage. The cleaned modeling dataset is included under `data/processed/`.


## Modeling Progression

1. **EDA + cleaning:** Filter raw NYC taxi data to Manhattan pickups and airport dropoffs.
2. **Regression:** Predict trip duration using baseline, linear regression, XGBoost, and Random Forest.
3. **Classification:** Define late trips as >120% of typical duration for similar trips and model late-risk probability.
4. **Decision layer:** Tune probability thresholds and create buffer-time guidance for travelers vs. operators.


## Final Results

| Task | Best model | Result | Why it matters |
|---|---|---:|---|
| Duration prediction | Random Forest / XGBoost | MAE ≈ 5.6 min | Converts noisy trips into useful ETA guidance |
| Late-risk classification | CatBoost | ROC-AUC ≈ 0.725 | Captures unusually slow trip risk under class imbalance |
| Risk-averse threshold | CatBoost @ 0.40 | Recall ≈ 0.826 | Catches most late trips for travelers |
| Baseline comparison | Historical median | MAE ≈ 13.1 min | ML cuts average error by more than half |
